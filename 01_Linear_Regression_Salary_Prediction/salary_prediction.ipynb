INDEPENDENT AND DEPENDENT FEATURES

```python
X=datasets
Y=df.target
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

FEATURE SCALING (STANDARIZING DATASET)

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
x_train=scaler.fit_transform(X_train)
```

```python
x_test=scaler.transform(X_test)
```

INTERCHANGE VALUE

```python
scaler.inverse_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
```

CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
```

MODEL TRAIN

```python
regression=LinearRegression()
regression.fit(x_train,y_train)

```

CROSS VALIDATION

```python
mse=cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

PREDICTION

```python
reg_predict=regression.predict(x_test)
```

```python
reg_predict
```

GRAPH

```python
import seaborn as sns
sns.distplot(reg_predict-y_test)
```

```python
df=pd.DataFrame({'Actual':y_test,'Predicted':reg_predict})
```

```python
df['Error']=df['Actual']-df['Predicted']
df['abs_error']=abs(df['Error'])
```

MEAN ABSOLUTE ERROR

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test,reg_predict))
```

MEAN SQUARED ERROR

```python
print(metrics.mean_squared_error(y_test,reg_predict))
```

ROOT MEAN SQUARED ERROR

```python
print(np.sqrt(metrics.mean_squared_error(y_test,reg_predict)))
```

GRAPH -> ACTUAL VS PREDICTED

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test,reg_predict)
plt.plot(y_test,y_test,'r')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
